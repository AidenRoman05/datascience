# Can you predict Boston's weather?
Many of you are new in town, so you may to know what the weather is going to be like in the following weeks. The dataset loaded below contains information about temperature highs, lows, and averages of temperature, dew point, wind speeds, sea level pressure, and precipitation levels for every day from 1/1/2013 - 4/8/2018 inclusive. All data is publicly available in its entirety on the Weather Underground website.

The goal, is to predict the weather in the upcoming weeks using a simple Markov chain. To do so we'll use the Pandas module: https://pandas.pydata.org

```python
def matrix_vector(m, v):
    r = []
    for row in m:
        x = sum([row[i] * v[i] for i in range(len(v))])
        r.append(x)
    return r
```

```python

def matrix_power(m, t):
    n = len(m)
    current_power = [[ x for x in row] for row in m]
    for i in range(t - 1):
        next_power = [[ 0 for i in range(n)] for i in range(n)]
        
        for j in range(n):
            
            for k in range(n):
                
                for l in range(n):
                    next_power[j][k] += m[j][l] * current_power[l][k]
        current_power = [[ x for x in row] for row in next_power]
    return current_power
```

```python
def discrete_ds(m, v, t):
    for i in range(t):
        v = matrix_vector(m, v)
    return v
```

```python
def unique_char(s):
    uniq = []
    for c in s:
        if c not in uniq:
            uniq.append(c)
    return len(uniq)
    
```

```python
def words(s):
    w = 1
    for c in s:
        if c == " ":
            w += 1
    return w
    
```

```python
def frequencies(s):
    freq = {}
    for c in s:
      if c in freq:
            freq[c] += 1
      else:
            freq[c] = 1
    return [[char, freq[char]] for char in freq]
```
